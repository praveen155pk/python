```python
# Program to print number of days in a given month
dict={'Jan':31,'Mar':31,'Apr':30,'May':31,'Jun':30,'Jul':31,'Aug':31,'Sep':30,'Oct':31,'Nov':30,'Dec':31}
month=input("Enter first 3 letters of month : ")
if month in dict:
    print("No of days in",month,"are",dict[month])
elif month=='Feb':
    year=int(input("Enter year : "))
    if year%4==0 and (year%100!=0 or year%400==0):
        print("No of days in Feb are 29")
    else:
        print("No of days in Feb are 28")
else:
    print("Enter correct correct")
```

```python
# Password encryption and decryption
password=input("Password : ")
encry=""
decry=""
for ch in password:
    encry=encry+chr(ord(ch)+3)
print("Password in Encrypted form is :",encry)
for ch in encry:
    decry=decry+chr(ord(ch)-3)
print("Password in Decrypted form is :",decry)
```

```python
# GPA calculation program
subjects=['CS   ','CO&A   ','OOP   ','FLAT   ','DBMS   ','E&HV   ']
sum=0
a=0
print("Enter marks for mentioned subjects below")
for i in subjects:
    marks=int(input(i))
    if marks in range(90,101):
        sum=sum+10
    elif marks in range(80,90):
        sum=sum+9
    elif marks in range(70,80):
        sum=sum+8
    elif marks in range(60,70):
        sum=sum+7
    elif marks in range(50,60):
        sum=sum+6
    elif marks in range(40,50):
        sum=sum+5
    elif marks in range(0,40):
        a=1
if a==1:
    print("\nFail")
else:
    print("\nGPA is",round(sum/6,2))
```

```python
# Word frequency count
sent=input("Enter any sentence : ")
word=input("Enter any word : ")
count=0
wl=len(word)
for ch in range(len(sent)-len(word)+1):
    if word==sent[ch:wl]:
        count+=1
    wl+=1
print(word,"is repeated",count,"times in the given sentence")
```

```python
# Matrix manipulation program
matrix1=[]
matrix2=[]
r=[]
n=int(input("How many number of rows are there for 2 square matrix : "))
print("Input numbers in 1st matrix :")
for i in range(n):
    for j in range(n):
        r.append(int(input()))
    matrix1.append(r)
    r=[]
print("Input numbers in 2nd matrix :")
for i in range(n):
    for j in range(n):
        r.append(int(input()))
    matrix2.append(r)
    r=[]
print("Entered matrices are")
for i in range(n):
    print(matrix1[i])
print()
for i in range(n):
    print(matrix2[i])
add=[]
for i in range(n):
    for j in range(n):
        r.append(matrix1[i][j]+matrix2[i][j])
    add.append(r)
    r=[]
print("\nAddition of above two matrices is :")
for i in range(n):
    print(add[i])
sub=[]
for i in range(n):
    for j in range(n):
        r.append(matrix1[i][j]-matrix2[i][j])
    sub.append(r)
    r=[]
print("\nSubtraction of above two matrices is :")
for i in range(n):
    print(sub[i])
mul=[]
a=0
for i in range(n):
    for j in range(n):
        for k in range(n):
            a+=matrix1[i][k]*matrix2[k][j]
        r.append(a)
        a=0
    mul.append(r)
    r=[]
print("\nMultiplication of above two matrices is :")
for i in range(n):
    print(mul[i])
```

```python
# Data transformation operations using pandas

import pandas as pd

# Creating a Dataframe
df=pd.DataFrame({"Regdno":[1,2,3,4,5],
                 "Name":['Abdal','Nithin','Praveen','Dileep','Ram'],
                 "Sub1":[10,20,30,40,50],
                 "Sub2":[89,85,89,84,79]})
rnames=['A','B','C','D','E']
df.index=rnames
print("\nOriginal Dataframe is : \n")
print(df)

# Performing Dataframe transformation operations

# Adding a new column to Dataframe
df['Sub3']=[45,56,67,78,89]
print("\n\nDataframe after adding new column Sub3 is : \n")
print(df)

# Adding a new row to Dataframe
df.loc['F',:]=[6,'Balaji',45,56,67]
print("\n\nDataframe after adding new row F is : \n")
print(df)

# Deleting a column and a row of Dataframe
df.drop('Sub3',axis=1,inplace=True)
df.drop('C',axis=0,inplace=True)
print("\n\nDataframe after deleting Sub3 column and C row is : \n")
print(df)

# Adding a column at specific position in Dataframe
df.insert(2,'Sub0',[23,34,45,56,67])
print("\n\nDataframe after adding new column Sub0 is : \n")
print(df)

# Adding a row at specific position in Dataframe
df=df.T # Transpose of df
df.insert(2,'G',[7,'Raja',99,98,97])
df=df.T # Transpose of df
print("\n\nDataframe after adding new row G is : \n")
print(df) 

# Adding 10 to specified column
df.loc[:,'Sub2']+=10
print("\n\nDataframe after adding 10 to data of Sub2 column is : \n")
print(df)
```

**While performing concatination of two tables row wise:**
    
    If index is ignored, every row will be given different index names in output.
    
    If index is not ignored, every row will be given the same index names in output as given while creating tables. 

**While performing concatination of two tables column wise:**
    
    If index is ignored, every column will be given different names but row indexes names will not be ignored. The output will be corresponding to their row indexes names.
    
    If index is not ignored, every column will be given same index names as given while creating tables but row indexes names will not be ignored. The output will be corresponding to their row indexes names.
